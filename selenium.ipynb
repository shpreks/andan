```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd


from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pickle
```

```python
service = Service(executable_path="C:/Users/persi/chromedriver.exe")  # указываем путь до драйвера
browser = webdriver.Chrome(service=service)
time.sleep(10)
```

```python
# сбор всех нужных страниц
top_leagues = ["GB1","ES1","IT1","L1","FR1","PO1","NL1","TR1","EB2","ES2","IT2","L2","FR2","PO2","NL2","TR2"]
url_league = "https://www.transfermarkt.com/premier-league/marktwerte/wettbewerb/"
global_soups = list()
# переход на страницу лиги
for league in top_leagues:
    browser.get(url_league + league)
    time.sleep(10)
    league_soups = list()
    # сбор данных со всех страниц лиги
    for i in range(2,5):
        try:
            time.sleep(10)
            league_soups.append(bs(browser.page_source, "html.parser"))
            browser.find_element(By.LINK_TEXT, str(i)).click()
        except Exception as ex:
            print(ex)
    global_soups.extend(league_soups)
```

```python
# поиск всех игроков

players = list()

for page in global_soups:
    page_players = page.select("td ~ .hauptlink > a") 
    page_players = [player["href"] for player in page_players][0::2] # футболисты на странице
    players.extend(page_players)
```

```python
len(players)
```

```python
with open('players_url.data', 'wb') as filehandle: # запись списка в файл
    pickle.dump(players, filehandle)  # данные с премьер лиги нормально не собираются
```

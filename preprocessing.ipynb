# Работа с выбросами / неверно собранными данными:

В сборе данных произошла ошибка, т.к. значение 0 для возраста, роста и стоимости игрока нереалистично.

```python
df.query("age == 0 | value == 0 | height == 0")
```

В датасете найден 21 игрок с подобной ошибкой. Т.к. данное количество мало по сравнению со всей выборкой, эти игроки удаляются из датасета.

```python
new_df = df[~(df["value"] == 0)]  # удаление строк с нулевым значением стоимости, т.к. они не несут практического смысла
```

В сборе данных явно произошла ошибка, т.к. максимальное значение прямых красных карточек превосходит значение вторых желтых в несколько раз, что не сопоставимо с реальностью, т.к. вряд ли один и тот же футболист может в почти каждой игре(судя по количеству красных карточек) получать прямую красную, и почти никогда не получать за фолы желтые(хотя судь редко сразу удаляют игрока даже за грубые фолы). К тому же такого футболиста очень быстро перестали бы выпускать из за зашкаливающего количества красных(78). К примеру у одного из самых грубых игроков(по мнению экспертов) Рамоса 27 красных карточек за всю карьеру. Следовательно в данных есть как минимум один выброс.

```python
new_df[new_df.red22 > 5].sort_values("red22")
```

Действительно, 64 игрока, судя по собранным данным, получили красных карточек больше чем 5, что является нереалистичным значением.
(у них также нереалистичные показатели красных карточек других годов)

```python
new_df[new_df.red22 < 5].red22.mean(), new_df[new_df.red22 < 5].red21.mean(), new_df[new_df.red22 < 5].red20.mean()
```

Вместо нереалистичных значений красных карточек для каждого года подставляется ноль, т.к. среднее красных карточек по каждому году для остальной выборки без таких выбросов близко к нулю.Т.е. с высокой вероятностью футболист не получает за сезон прямых красных и это преобразование не отразится на качестве данных.

```python
new_df.loc[new_df["red22"] > 5, "red20"] = 0
new_df.loc[new_df["red22"] > 5, "red21"] = 0
new_df.loc[new_df["red22"] > 5, "red22"] = 0
```

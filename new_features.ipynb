# Создание новых параметров: 
__Среднее количество ассистов и голов за матч__. Футболист, сыгравший меньше матчей, обладает меньшей вероятностью забить большое количество голов чем тот, кто сыграл больше. Поэтому важно посмотреть сколько в среднем за матч забивает футболист, т.к. это нивелирует влияние общего количества матчей на результативность.(аналогичная логика и с ассистами)

```python
for i in range(20, 23):
    df["goal_percentage" + str(i)] = np.where(df['matches' + str(i)]>0, df["goals" + str(i)] / df["matches" + str(i)], 0)
```

```python
for i in range(20, 23):
    df["assists_percentage" + str(i)] = np.where(df['matches' + str(i)]>0,df["assists" + str(i)] / df["matches" + str(i)], 0)
```

__Обобщенные позиции на поле.__ Слишком много значений у переменной позиция, большинство этих значений принадлежат малому количеству футболистов. Значения данной категории обьединяются в 4 обобщенные группы(голкипер, защитники, полузащитники, нападающие). Создается 4 дополнительных колонки(голкипер, защитники, полузащитники, нападающие), если футболист принадлежит к обобщенной категории, значение столбца 1 и наоборот.

```python
df.position.nunique()
```

```python
goalkeeper = ["Goalkeeper"]
deff = ["Centre-Back", "Right-Back", "Left-Back"]
middle = ["Attacking Midfield", "Defensive Midfield", "Central Midfield", "Right Midfield", "Left Midfield"]
attack = ["Centre-Forward", "Right Winger", "Left Winger", "Second Striker"]
```

```python
df["goalkeeper"] = df["position"].isin(goalkeeper) * 1
df["deff"] = df["position"].isin(deff) * 1
df["middle"] = df["position"].isin(middle) * 1
df["attack"] = df["position"].isin(attack) * 1
```
